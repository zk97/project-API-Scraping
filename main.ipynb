```python
import asyncio
import re
import time
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import aiohttp

class ProjectMainSpider:
    """
    This is the constructor class to which you can pass a bunch of parameters. 
    These parameters are stored to the class instance variables so that the
    class functions can access them later.
    
    url_pattern: the regex pattern of the web urls to scape
    pages_to_scrape: how many pages to scrape
    sleep_interval: the time interval in seconds to delay between requests. If <0, requests will not be delayed.
    content_parser: a function reference that will extract the intended info from the scraped content.
    """
    def __init__(self, url_pattern, pages_to_scrape=10, sleep_interval=True, get_agent=None):
        self.url_pattern = url_pattern
        self.pages_to_scrape = pages_to_scrape
        self.sleep_interval = sleep_interval
        self.get_agent=get_agent
        self.content_parser = None
        self.links=[]
        self.names=[]
        self.brands=[]
        self.ratings=[]
        self.price=[]

    async def fetch_headers(self,url,headers):
        async with aiohttp.ClientSession() as s, s.get(url,headers=headers) as res:
            ret=await res.read()
            status=res.status
            return status,ret
    """
    Scrape the content of a single url.
    """
    
    async def scrape_url(self,url):
        agent=self.get_agent()
        headers = {
            'user-agent': agent,
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en;q=0.9,es;q=0.8,fr;q=0.7,gl;q=0.6',
        } 
        try:
            status,content= await self.fetch_headers(url,headers)
            if status<300:                    
                result = self.content_parser(content)
            elif status in range(400,500):
                print('Error {} in "{}", request failed because the resource either does not exist or is forbidden.'
                      .format(status,url))
            else:
                print('Error {}, in "{}", request failed because the response server encountered an error.'
                      .format(status,url))
        except aiohttp.ServerTimeoutError:
            print('Error= Timeout.'.format(url))
        except aiohttp.TooManyRedirects:
            print('Error= TooManyRedirects, in {}.'.format(url))
        except aiohttp.ClientSSLError:
            print('Error= SSLError, in {}.'.format(url))
        except aiohttp.ClientError as e:
            print('Error= {}, in {}.'.format(e,url))
    
    def random_interval(self):
        delays = [1, 2, 3, 4, 5, 7]
        perm = np.random.RandomState()
        #return random x in [0.5,n+0.5]
        return(perm.permutation(range(np.random.choice(delays)*100))[0]+5)/100
    """
    After the class is instantiated, call this function to start the scraping jobs.
    This function uses a FOR loop to call `scrape_url()` for each url to scrape.
    """
    async def kickstart(self,parser,main=True):
        self.content_parser=parser
        #Get flower links from catalog
        if main:
            #send requests in n groups of n where n=sqrt(total requests)
            for j in range(int(np.sqrt(self.pages_to_scrape))):
                await asyncio.wait([self.scrape_url(self.url_pattern % i) for i in range(1, self.pages_to_scrape+1)
                                   if i%(int(np.sqrt(self.pages_to_scrape)))==j])
                if self.sleep_interval:
                    await asyncio.sleep(self.random_interval())
                #Progress indicator
                print('Group {} finished from {}'.format(j+1,int(np.sqrt(self.pages_to_scrape))))
        #Get info from each flower page
        else:
            #send requests in n groups of n where n=sqrt(total requests)
            for j in range(int(np.sqrt(len(self.links)))):
                await asyncio.wait([self.scrape_url(i) for x,i in enumerate(self.links) 
                                    if x%(int(np.sqrt(len(self.links))))==j])
                if self.sleep_interval:
                    await asyncio.sleep(self.random_interval())
                    #Progress indicator
                print('Group {} finished from {}'.format(j+1,int(np.sqrt(len(self.links)))))
            #Save data in DF
            data=zip(self.names,self.brands,self.ratings,self.price)
            self.df=pd.DataFrame(data,columns=['Name','Brand','Rating','Price'])
        print('Done')
            
    def async_kickstart(self,parser,main=True):
        try:
            loop=asyncio.get_event_loop()
            loop.run_until_complete(self.kickstart(parser,main))
        except:
            pass
        
                
    def main_flower_parser(self,content):
        soup=BeautifulSoup(content)
        self.links+=['https://www.leafly.com'+x['href'] for x in soup.select('a.flex')]
        
    def flower_parser(self,content):
        soup=BeautifulSoup(content)
        name=[x.h1.text for x in soup.select('.cp-product-details__wrap.mb-lg')]
        brand=[x.h2.a.text for x in soup.select('.cp-product-details__wrap.mb-lg')]
        ratings=[x.text for x in soup.select('[itemprop="ratingValue"]:not(.star-rating)')]
        price=[x.text+'.'+y.text for y in soup.select('.money-decimal') for x in soup.select('.money-whole')]
        for i in [name,brand,ratings,price]:
            if not len(i):
                i.append(None)
        self.names+=name
        self.brands+=brand
        self.ratings+=ratings
        self.price+=price

def random_agent():
    with open('agents.txt',encoding="utf8") as f:
        lines = f.readlines()
        perm = np.random.RandomState()
        index = perm.permutation(len(lines) - 1)[0]
        agent = lines[int(index)].strip()
        return agent
```

```python
import asyncio
import re
import time
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import aiohttp

class IronhackStrainSpider:
    """
    This is the constructor class to which you can pass a bunch of parameters. 
    These parameters are stored to the class instance variables so that the
    class functions can access them later.
    
    url_pattern: the regex pattern of the web urls to scape
    pages_to_scrape: how many pages to scrape
    sleep_interval: the time interval in seconds to delay between requests. If <0, requests will not be delayed.
    content_parser: a function reference that will extract the intended info from the scraped content.
    """
    def __init__(self, url_pattern, pages_to_scrape=10, sleep_interval=True, get_agent=None):
        self.url_pattern = url_pattern
        self.pages_to_scrape = pages_to_scrape
        self.sleep_interval = sleep_interval
        self.get_agent=get_agent
        self.content_parser = None
        self.strain_links=[]
        self.strain_names=[]
        self.strain_dominances=[]
        self.strain_thc=[]
        self.strain_cbd=[]
        self.strain_terpenes=[]
        self.strain_more=[]
        self.strain_calm_energ=[]
        self.strain_eff_feel=[]
        self.strain_eff_help=[]
        self.strain_eff_negative=[]
        
    async def fetch_headers(self,url,headers):
        async with aiohttp.ClientSession() as s, s.get(url,headers=headers) as res:
            ret=await res.read()
            status=res.status
            return status,ret
        
    """
    Scrape the content of a single url.
    """
    async def scrape_url(self,url):
        agent=self.get_agent()
        headers = {
            'user-agent': agent,
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en;q=0.9,es;q=0.8,fr;q=0.7,gl;q=0.6',
        } 
        try:
            status,content= await self.fetch_headers(url,headers)
            if status<300:                    
                result = self.content_parser(content)
            elif status in range(400,500):
                print('Error {} in "{}", request failed because the resource either does not exist or is forbidden.'
                      .format(status,url))
            else:
                print('Error {}, in "{}", request failed because the response server encountered an error.'
                      .format(status,url))
        except aiohttp.ServerTimeoutError:
            print('Error= Timeout.'.format(url))
        except aiohttp.TooManyRedirects:
            print('Error= TooManyRedirects, in {}.'.format(url))
        except aiohttp.ClientSSLError:
            print('Error= SSLError, in {}.'.format(url))
        except aiohttp.ClientError as e:
            print('Error= {}, in {}.'.format(e,url))
    
    def random_interval(self):
        delays = [7, 4, 6, 2, 10, 19]
        perm = np.random.RandomState()
        return(perm.permutation(range(np.random.choice(delays)*100))[0]+50)/100
    """
    After the class is instantiated, call this function to start the scraping jobs.
    This function uses a FOR loop to call `scrape_url()` for each url to scrape.
    """
    async def kickstart(self,parser,main=True):
        self.content_parser=parser
        interval=0
        #Get strain links from cataloge
        if main:
            #send requests in n groups of size n where n=sqrt(total requests)
            for j in range(int(np.sqrt(self.pages_to_scrape))):
                await asyncio.wait([self.scrape_url(self.url_pattern % i) for i in range(1, self.pages_to_scrape+1)
                                   if i%(int(np.sqrt(self.pages_to_scrape)))==j])
                if self.sleep_interval:
                    await asyncio.sleep(self.random_interval())
                print('Group {} finished from {}'.format(j+1,int(np.sqrt(self.pages_to_scrape))))
        #Get strains data
        else:
            #send requests in 2n groups of size n/2 where n=sqrt(total requests)
            for j in range(int(np.sqrt(len(self.strain_links)))*2):
                await asyncio.wait([self.scrape_url(i) for x,i in enumerate(self.strain_links) 
                                    if x%(int(np.sqrt(len(self.strain_links)))*2)==j])
                if self.sleep_interval:
                    await asyncio.sleep(self.random_interval())
                print('Group {} finished from {}'.format(j+1,int(np.sqrt(len(self.strain_links)))*2))
            data=zip(self.strain_names,self.strain_dominances,self.strain_thc,self.strain_cbd,self.strain_terpenes,self.strain_more,
                  self.strain_calm_energ,self.strain_eff_feel,self.strain_eff_help,self.strain_eff_negative)
            self.df=pd.DataFrame(data,columns=['Name','Dominance','THC','CBD','Terpenes','More','Calm/Energize','Feelings','Helps','Negative'])
        print('Done')

    def async_kickstart(self,parser,main=True):
        try:
            loop=asyncio.get_event_loop()
            loop.run_until_complete(self.kickstart(parser,main))
        except:
            pass
        
    def main_strain_parser(self,content):
        soup=BeautifulSoup(content)
        self.strain_links+=['https://www.leafly.com'+x['href']+'/info' for x in soup.select('a.strain-tile')]
        
    
    def strain_parser(self,content):
        soup=BeautifulSoup(content)
        panels=soup.select('div[role="tabpanel"]')
        name=[x.text for x in soup.select('h1.text-xl')]
        dominance=[x.text for x in soup.select('h1.text-xl+div')]
        thc=[x.text for x in panels[0].select('div.font-bold')]
        cbd=[x.text for x in panels[1].select('div.font-bold')]
        terpenes=[[x.text for x in panels[2].select('div.font-bold')]]
        more_name=[x.text for x in panels[3].select('div.font-bold')]
        more_quants=[x.text for x in panels[3].select('div.font-bold+div')]
        more=[list(zip(more_name,more_quants))]
        #Prevents index out of bounds if empty
        if len(soup.select('.calm-energize__mark'))>0:
            #Save calm_energ info in [-1,1] inverval
            calm_energ=[round((float(x)-50)/50,4) for x in re.findall(':(.*)%',soup.select('.calm-energize__mark')[0]['style'])]
        else:
            calm_energ=[]
        #Prevents index out of bounds if empty
        if len(soup.select('div.react-tabs'))>0:
            effects=[x for x in soup.select('div.react-tabs')[0].div]
            effects_feelings=[[x.text for x in effects[0]]]
            effects_helps=[[x.text for x in effects[1]]]
            effects_negative=[[x.text for x in effects[2]]]
        else:
            effects_feelings=[]
            effects_helps=[]
            effects_negative=[]
        #Preserves order in data 
        for i in [name,dominance,thc,cbd,terpenes,more,calm_energ,effects_feelings,effects_helps,effects_negative]:
            if not len(i):
                i.append(None)
        self.strain_names+=name
        self.strain_dominances+=dominance
        self.strain_thc+=thc
        self.strain_cbd+=cbd
        self.strain_terpenes+=terpenes
        self.strain_more+=more
        self.strain_calm_energ+=calm_energ
        self.strain_eff_feel+=effects_feelings
        self.strain_eff_help+=effects_helps
        self.strain_eff_negative+=effects_negative
    

def random_agent():
    with open('agents.txt',encoding="utf8") as f:
        lines = f.readlines()
        perm = np.random.RandomState()
        index = perm.permutation(len(lines) - 1)[0]
        agent = lines[int(index)].strip()
        return agent
```
